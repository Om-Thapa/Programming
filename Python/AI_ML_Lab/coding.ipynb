```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 12))

x_exp = np.linspace(-10, 1)
plt.subplot(3, 3, 1)
plt.plot(x_exp, np.exp(x_exp))

step = np.pi / 15
x_trig = np.arange(-4 * np.pi, 4 * np.pi + step/2, step)
plt.subplot(3, 3, 2)
plt.plot(x_trig, np.sin(x_trig))
plt.axhline(0, color="black", linewidth=1)

plt.axhline(0,color="black",linewidth=1)

plt.show()
```
