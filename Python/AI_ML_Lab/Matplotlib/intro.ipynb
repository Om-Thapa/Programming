```python
# Generate 100 sample data with values from 0 to 2
x = np.linspace(0, 2, 100)

# Default size in inches
# Contrained layout minimizes clashed and overlap, enduringm all plot decorations fits within the figure.
fig, ax = plt.subplots( figsize=(5, 3), layout = 'constrained')
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
```

```python
def my_plotter(ax, data1, data2, param_dict) : 
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (5, 2.5))
my_plotter(ax1, data1, data2, {'marker':'x'})
my_plotter(ax2, data3, data4, {'marker':'o'})
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(5, 2.5), layout='constrained')
x_axis_data = np.arange(len(data1))
data = 10**data1
axs[0].plot(x_axis_data, data)
axs[1].set_yscale('log')
axs[1].plot(x_axis_data, data)
```

```python
fig, axs = plt.subplots(2,1, layout='constrained')
axs[0].plot(x_axis_data, data)
axs[0].set_title('Automatic ticks')

axs[1].plot(x_axis_data, data)
axs[1].set_xticks(np.arange(0, 100, 30), ['aero', '30', 'sixty', '90'])
axs[1].set_yticks([0, 40, 80, 120, 160])
axs[1].set_title('Manual ticks')
```

```python
X, Y = np.meshgrid( np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig, axs = plt.subplots(figsize = (4, 2.7), layout='constrained')
co = axs.contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs)
axs.set_title('contoutf()')
plt.show()
```
